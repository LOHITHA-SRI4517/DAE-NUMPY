```python
import numpy as np

roll = [1,2,3,4,5]

attendance  = np.zeros(len(roll),dtype = int)
# print(attendance)
present = np.ones(len(roll),dtype = int)

for i in range (len(roll)):
    res = input(f"Roll Number : {roll[i]} P/A :")
    if res.upper() == "P":
      attendance[i] = present[i]
    else:
      attendance[i] = 0

for i in range(len(roll)):
  print(f"Roll Number : {roll[i]} : {attendance[i]}")
```
