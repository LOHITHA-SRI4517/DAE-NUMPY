4. Size Function


    *In python numpy.size() function counts the no. of elements alog the axis (0-Rows,1-Columns)*


```python
pip install numpy
```

```python
import numpy as np
arr=np.array([[1,2,3], [4,5,6], [10,20,30], [40,50,60]])
print(arr)
print("TOTAL : ",arr.size)
print("ROWS : ",np.size(arr,0))
print("COLUMNS :",np.size(arr,1))
```
