## **Sorting a 2D NumPy Array**

When sorting a 2D array, you can specify the `axis` along which to sort:

*   `axis=0`: Sorts along the columns (each column is sorted independently).
*   `axis=1`: Sorts along the rows (each row is sorted independently).
*   `axis=None` (default): Flattens the array first, then sorts it as a 1D array.

```python
import numpy as np

arr_2d = np.array([[3, 1, 4],
                   [6, 5, 2],
                   [9, 7, 8]])

print("Original 2D Array:")
print(arr_2d)

print("\nSorted along axis=0 (columns):")
print(np.sort(arr_2d, axis=0))

print("\nSorted along axis=1 (rows):")
print(np.sort(arr_2d, axis=1))

print("\nSorted after flattening (axis=None):")
print(np.sort(arr_2d, axis=None))
```
